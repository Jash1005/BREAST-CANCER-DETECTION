```python
information = pd.read_csv('Breast Cancer data/Info.txt',delim_whitespace=True) # delimwhitespace = any amt of space is colm seperator

information.head(2)
```

```python
information['PGM'] = information['REFNUM'] + '.pgm'

information['path'] = 'Breast Cancer data/Images/' + information['PGM']

information.head(2)
```

```python
# Severity means how serious this abnormality is => converting into integer for better understanding

print(information['SEVERITY'].unique())

information['label_of_severity'] = information['SEVERITY'].map({'B' : 1, 'M' : 2})  # 0 : normal
information['label_of_severity'] = information['label_of_severity'].fillna(0).astype(int)
information.head(2)
```

```python
import os
from PIL import Image
import cv2
import numpy as np

def crop_with_mask(image, mask):
    """
    Crop the image to the bounding box of the non-zero region in the mask.
    """
    coords = cv2.findNonZero(mask.astype(np.uint8))  # Find all non-zero points
    x, y, w, h = cv2.boundingRect(coords)  # Get bounding box
    cropped = image[y:y+h, x:x+w]  # Crop image to bounding box
    return cropped



count = 1
output_dir = "Breast Cancer data/Cropped Images"
os.makedirs(output_dir, exist_ok=True)

for path in information['path']:
    # Load image in grayscale
    image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

    if image is None:
        print(f"Failed to load: {path}")
        continue

    # Generate binary mask
    _, mask = cv2.threshold(image, 10, 1, cv2.THRESH_BINARY)

    # Crop using mask
    cropped = crop_with_mask(image, mask)

    # Convert to PIL image
    img = Image.fromarray(cropped)

    # Save image with proper filename
    filename = f"mdb{count:03d}.png"  # Ensures names like mdb001.png
    save_path = os.path.join(output_dir, filename)
    img.save(save_path)
    print(f"Saved: {save_path}")

    count += 1
```

```python
from PIL import Image
import numpy as np

def convert_Image(path, size=(224, 224)):
    img = Image.open(path).convert("L")  # Convert to grayscale
    img = img.resize(size)
    arr = np.array(img) / 255.0  # Normalize to [0, 1]
    return arr
```

```python
# Assuming all cropped images were saved using this naming pattern:
cropped_paths = [f"Breast Cancer data/Cropped Images/mdb{idx+1:03d}.png" for idx in range(len(information))]
information['cropped_image'] = cropped_paths

# Convert image to array (define convert_Image first!)
X = np.array([convert_Image(p) for p in information['cropped_image']])

# Get labels
Y = information['label_of_severity'].values
```

```python
# now creating Training and Testing data

from sklearn.model_selection import train_test_split

X_train , X_test , Y_train , Y_test  =  train_test_split(X,Y,test_size = 0.2,random_state = 42)


print(X_train[0])
print(Y_train[0])
```

```python
X_train = X_train.reshape(-1, 224, 224, 1).astype('float32')

X_test = X_test.reshape(-1, 224, 224, 1).astype('float32')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense

model = Sequential([
    Input(shape=(224, 224, 1)),
    Conv2D(32, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(3, activation='softmax')  # 3-class output
])
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
import numpy as np
from tensorflow.keras.utils import to_categorical

# Ensure labels are flat, not one-hot or multi-dim
Y_train = np.array(Y_train).flatten()
Y_test = np.array(Y_test).flatten()

# Encode only if they're not already one-hot
if Y_train.ndim == 1:
    Y_train = to_categorical(Y_train, num_classes=3)
    Y_test = to_categorical(Y_test, num_classes=3)

# Train the model
model.fit(X_train, Y_train, epochs=10, validation_data=(X_test, Y_test), batch_size=16)
```

```python
# Get predicted probabilities
pred_probs = model.predict(X_test)

# Convert probabilities to class labels
predicted_classes = np.argmax(pred_probs, axis=1)

label_map = {0: "NORMAL", 1: "BENIGN", 2: "MALIGNANT"}

readable_preds = [label_map[i] for i in predicted_classes]
```

```python
from sklearn.metrics import classification_report, accuracy_score

true_classes = np.argmax(Y_test, axis=1)  # if one-hot encoded
print("Accuracy:", accuracy_score(true_classes, predicted_classes))
print(classification_report(true_classes, predicted_classes, target_names=label_map.values()))
```
